```python
import boto3

def create_or_update_dashboard(
    aws_account_id,
    dashboard_id,
    dashboard_name,
    analysis_arn,
):
    client = boto3.client("quicksight")

    try:
        return client.create_dashboard(
            AwsAccountId=aws_account_id,
            DashboardId=dashboard_id,
            Name=dashboard_name,
            SourceEntity={
                "SourceAnalysis": {
                    "Arn": analysis_arn,
                    "DataSetReferences": []
                }
            }
        )
    except client.exceptions.ResourceExistsException:
        return client.update_dashboard(
            AwsAccountId=aws_account_id,
            DashboardId=dashboard_id,
            Name=dashboard_name,
            SourceEntity={
                "SourceAnalysis": {
                    "Arn": analysis_arn,
                    "DataSetReferences": []
                }
            }
        )
```
