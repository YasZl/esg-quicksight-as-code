## 1. Setup & Imports

```python
# Add parent directory to path for imports
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))

# Import our dashboard renderer
from esg_lib.dashboard_renderer import (
    load_portfolio_data,
    create_kpi_card,
    create_security_type_pie,
    create_top_holdings_bar,
    create_dashboard_html
)

import pandas as pd
import plotly.io as pio

# Set plotly renderer for Jupyter
pio.renderers.default = 'notebook'

print("✅ Imports successful!")
```

## 2. Load Data

```python
# Path to data file
data_file = "../demo_cross_asset_type_2025-03-31_ORIGINAL.xlsx"

# Load portfolio data
df = load_portfolio_data(data_file)

print(f"📊 Data loaded successfully!")
print(f"   - Rows: {len(df):,}")
print(f"   - Columns: {len(df.columns)}")
print(f"   - Portfolio Date: {df['Portfolio date'].iloc[0]}")
print(f"\n📋 First 3 rows:")
df.head(3)
```

## 3. Data Summary

```python
# Calculate key statistics
total_value = df['Security market valuation in portfolio currency'].sum()
security_type_dist = df.groupby('Security type')['Security market valuation in portfolio currency'].sum()

print("💰 Portfolio Statistics:")
print(f"   - Total Invested Value: €{total_value:,.0f}")
print(f"   - Number of Holdings: {len(df):,}")
print(f"   - Average Holding Size: €{total_value/len(df):,.0f}")
print(f"\n🏢 Security Type Distribution:")
for sec_type, value in security_type_dist.items():
    pct = (value / total_value) * 100
    print(f"   - {sec_type}: €{value:,.0f} ({pct:.1f}%)")
```

## 4. Visualizations

### 4.1 KPI Card - Total Invested Value

```python
fig_kpi = create_kpi_card(df)
fig_kpi.show()
```

### 4.2 Pie Chart - Investment Distribution by Security Type

```python
fig_pie = create_security_type_pie(df)
fig_pie.show()
```

### 4.3 Bar Chart - Top 10 Holdings

## 6. Additional Analysis (Optional)

### Top Holdings Details

```python
# Show top 10 holdings with details
top_10 = df.nlargest(10, 'Security market valuation in portfolio currency')[[
    'Security name',
    'Security type',
    'Security market valuation in portfolio currency'
]].copy()

top_10.columns = ['Security Name', 'Type', 'Value (€)']
top_10['Value (€)'] = top_10['Value (€)'].apply(lambda x: f"€{x:,.0f}")
top_10.reset_index(drop=True, inplace=True)
top_10.index += 1

print("🏆 Top 10 Holdings:")
top_10
```
